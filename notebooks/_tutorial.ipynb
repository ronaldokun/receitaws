```python
corp = pd.read_csv('dados/CadastroSRF_202209301224.csv', dtype={'NumCNPJCPF': 'string'})
corp
```

```python
corp['Tamanho'] = corp.NumCNPJCPF.map(len)
corp.loc[corp.Tamanho == 11, 'Tipo'] = 'CPF'
corp.loc[corp.Tamanho == 14, 'Tipo'] = 'CNPJ'
```

```python
corp.drop('Tamanho', axis=1, inplace=True)
```

```python
corp.to_excel('dados/Base_CORPORATIVO_PD_DataCarga_6meses.xlsx')
```

```python
corp_cpf = corp[corp.Tipo == 'CPF'].reset_index(drop=True)
corp_cnpj = corp[corp.Tipo == 'CNPJ'].reset_index(drop=True)
```

```python
cpf.CNPJ_CPF.isin(corp_cpf.NumCNPJCPF).sum()
```

```python
cpf['Cache_6m'] = False
cpf.loc[cpf.CNPJ_CPF.isin(set(corp_cpf.NumCNPJCPF.tolist())), 'Cache_6m'] = True
cpf.Cache_6m.value_counts()
```

```python
cnpj['Cache_6m'] = False
cnpj.loc[cnpj.CNPJ_CPF.isin(set(corp_cnpj.NumCNPJCPF.tolist())), 'Cache_6m'] = True
cnpj.Cache_6m.value_counts()
```
