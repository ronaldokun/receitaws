```python
from validate_docbr import CPF, CNPJ
from receitaws.consultas import Requisicao, requisitar_em_lote
import pandas as pd
from fastcore.xtras import Path
```

```python
entrada = Path.cwd().parent / 'dados/CPF_AFFO.xlsx'
```

```python
cpf = pd.read_excel(entrada, dtype='string')
```

```python
cpf
```

```python
cpf_in = cpf[cpf.cpf.isna()].reset_index(drop=True)
cpf_in = cpf_in.dropna(axis=1)
```

```python
cpf_in.shape
```

```python
cpf_in
```

```python
cpf_usuario = input('Digite o CPF do solicitante: ')
origem = "GIIB4 - Atualização AFFO"
ambiente = 'pd'
tipo = 'cpf'
```

```python
req = Requisicao(cpf_usuario, tipo, ambiente, origem, cache=12)
```

```python
cpfs = req.validar_identificadores(cpf_in.CNPJ_CPF.tolist())
```

```python
cpfs
```

```python
saida = Path.cwd().parent / 'dados/CPF_AFFO_20230207.xlsx'
```

```python
dados_cpf = requisitar_em_lote(cpfs, cpf_usuario, tipo, origem, ambiente, cache=6, saida=saida, n_workers=2)
```

```python
dados_cpf
```

```python
cpf_in = cpf_in.merge(dados_cpf, left_on='CNPJ_CPF', right_on='cpf', how='left')
```

```python
cpf_in
```

```python
pd.concat([cpf[cpf.cpf.notna()], cpf_in]).to_excel(entrada, index=False)
```
