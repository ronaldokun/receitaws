```python
from validate_docbr import CPF, CNPJ
from receitaws.consultas import Requisicao, requisitar_em_lote
import pandas as pd
import numpy as np
from fastcore.xtras import Path
from datetime import datetime
```

```python
pasta = Path.cwd().parent / 'dados'
file  = pasta / 'mala_direta_outorgadas.xlsx'
```

```python
df = pd.read_excel(file, dtype = 'string')
df
```

```python
df.loc[df['Data Consulta Receita'] == '-', 'Data Consulta Receita']
```

```python
df.loc[df['Data Consulta Receita'] == '-', 'Data Consulta Receita'] = '2019-01-01 00:00:00.000000'
```

```python
df['Data Consulta Receita'] = df['Data Consulta Receita'].apply(np.datetime64)
```

```python
df.sort_values('Data Consulta Receita', inplace=True)
```

```python
df = df.reset_index(drop=True)
```

```python
cpf = df[df['Natureza da Entidade'] == 'CPF']
```

```python
cnpj = df[df['Natureza da Entidade'] == 'CNPJ']
```

```python
subset_cpf = cpf[cpf.index < 18000]
subset_cpf
```

```python
subset_cnpj = cnpj[cnpj.index < 18000]
subset_cnpj
```

```python
cpf_usuario = input('Digite o CPF do solicitante: ')
origem = "Requisicao_Mala_Direta_Outorgadas"
ambiente = 'pd'
tipo = 'cnpj'
```

```python
req = Requisicao(cpf_usuario, tipo, ambiente, origem, cache=6)
```

```python
saida = Path.cwd().parent / 'dados/resultados_mala_direta_cnpj_pd.xlsx'
```

```python
dados_cnpj = requisitar_em_lote(subset_cnpj.loc[:, 'CPF_CNPJ Entidade Representada'].tolist(), cpf_usuario, tipo, origem, ambiente, cache=6, saida=saida, n_workers=1)
```

```python
dados_cnpj
```

```python
dados_cnpj = dados_cnpj.fillna('')
```

```python
cnpj_out = pd.merge(cnpj, dados_cnpj, left_on='CPF_CNPJ Entidade Representada', right_on='cnpj', how='left').fillna('')
```

```python
saida = Path.cwd().parent / 'dados/resultados_mala_direta_cnpj_pd_mesclados.xlsx'
cnpj_out.to_excel(saida, index=False)
```
